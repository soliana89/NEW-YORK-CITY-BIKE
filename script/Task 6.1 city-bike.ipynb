### findings : no duplicates exist 

```python
# Checking for Mixed-Data Types; 

for col in df.columns.tolist():
  weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

###  findings : no mixed data type

##  step 3.change performed 

```python
#defining 'gender' type characterstics;
Rider_Gender = []

for value in df["gender"]:
    if value == 1:
        Rider_Gender.append("Male")
    elif value == 2:
        Rider_Gender.append("Female")
    else:
        Rider_Gender.append("Unknown")
  
```

```python
df['Rider_Gender'] = Rider_Gender
```

```python
df['Rider_Gender'].value_counts(dropna = False)
```

```python
# cheking for new column;
df.head()
```

```python
# deleting unnecessary column;
# 'trip_ID' and 'birth_year' will not be needed for this analysis
df = df.drop(columns = ['trip_id', 'birth_year',])
```
